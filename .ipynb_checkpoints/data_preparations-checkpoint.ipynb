```python
def clean_column_names(df):
    # Clean column names
    cleaned_columns = ['_'.join(col.strip().split()).lower() for col in df.columns]
    cleaned_columns = [ i.replace('//','/') for i in cleaned_columns]
    df.columns = cleaned_columns
    return df

df = clean_column_names(df)
```

# Rename by Add Substring to df Columns Names

```python
def rename_colnames(df, substring):
    
    df_copy = df.copy()
    for col in df.columns:
        new_col_name = col + substring
        df_copy.rename(columns={col: new_col_name}, inplace=True)

    return(df_copy)


# df1 = rename_colnames(df, '_rm_by_iqr')
```

# Outlier Detection & Removal

```python
# remove outliers per each numerical col using either IQR, std or zcore
def remove_outliers(df, method):
    
    df = df.select_dtypes(include=['int', 'float'])    
    
    outliers_removed = pd.DataFrame()

    for col in df.columns:
        stats = df[col].describe()
        mean = stats.loc['mean']
#         min_val = stats.loc['min']
#         max_val = stats.loc['max']
        
        if method == 'IQR':
            Q1 = stats.loc['25%']
            Q3 = stats.loc['75%']
            IQR = Q3 - Q1
            lower_threshold = Q1 - 1.5 * IQR
            upper_threshold = Q3 + 1.5 * IQR

            # Remove outliers by std method:
            outliers_removed[col] = df[col][(df[col] >= lower_threshold) & (df[col] <= upper_threshold)]

        elif method == 'std':
            std = stats.loc['std']
            threshold = 3 * std
            outliers_removed[col] = df[col][abs(df[col] - mean) <= threshold]

        elif method == 'zscore':
            z_scores = zscore(df[col])
            threshold = 3
            outliers_removed[col] = df[col][abs(z_scores) <= threshold]
            
        else:
            next

    return outliers_removed


# df4 = remove_outliers(df4, method = 'IQR')
# df5 = remove_outliers(df5, method = 'std')
# df6 = remove_outliers(df6, method = 'zscore')
```

# Handling Missing Values - Imputation
### Inputation of randomsampling, mean, median, mode (most-freq), knn, regression

```python
#imputation
# to replace nan in num by random sampling of its own column
# Define a function for random sampling imputation

np.random.seed(123)
pd.options.mode.chained_assignment = None  # default='warn'
# pd.options.mode.chained_assignment = 'warn'


def random_sampling_imputation(df, columns):
    # Get non-missing values per col
    for col in columns:
        non_missing = df[col].dropna()
        # Generate random indices
        rand_index = np.random.choice(non_missing.index, size=df[col].isnull().sum(), replace=True)
         # Replace missing values with randomly sampled index's values of same col
        df[col].iloc[df[col].isnull()] = non_missing[rand_index].values
        # this part help to quickly get the random sample index used for imputation, validation
        # col1 = [int(str(x).replace(' ',',')) for x in rand_index]
        # # col = [int(x) for x in col]
        # print(col1)        
    return df


# impute_col = ['discounts','profit']
# df = random_sampling_imputation(df, impute_col)
```

# Handle Skewness (Transformation w/ Log, Boxcox or YeoJohnson)
### incorporate 3 methods for now; log for non-zero, boxcox for all positive, and yeojohnson for everything else (-0+)

```python
# Shapiro-Wilk test for normality
# Normality test:
# Null hypothesis (H0): The data follows a normal distribution.
# Alternative hypothesis (H1): The data does not follow a normal distribution.
# Box-Cox transformation is a more generalized approach that can handle both positive and negative skewness.It requires the data to be strictly positive.
# Yeo-Johnson transformation is an extension of the Box-Cox transformation that can handle both positive and negative skewness, including zero and negative values.

def handle_skewness(df):
    
    numerical_cols = df.select_dtypes(include=['number']).columns
    # print(numerical_cols)
    for col in numerical_cols:
        statistic, p_value = stats.shapiro(df[col])        
        if p_value >= 0.05:
            print(col, ' is normally distributed.')
            next
            
        else:
            print("statistic: ", statistic, "p_value: ", p_value)
            plt.figure(figsize=(4, 2))  # Adjust the figsize as needed
            stats.probplot(df[col], dist="norm", plot=plt)
            plt.title("Q-Q Plot for col: " + col)
            plt.xlabel("Theoretical Quantiles")
            plt.ylabel("Sample Quantiles")
            plt.show()
            # plt.figure(figsize=(4,2))
            # sns.histplot(df[col], kde=True)
            # plt.title("Histogram of col: " + col)
            # plt.xlabel("Values")
            # plt.ylabel("Frequency")
            # plt.show()
            
            log_trans = boxcox_trans = yeojohn_trans = None
            # only log transform when no zero in array
            if (df[col] == 0).sum() ==0:  
                log_trans = np.log(df[col])
                print('log trans')
            else:
                # boxcox transform for all positive values
                if (np.all(df[col] >0)):
                    boxcox_trans, lambda_ = boxcox(df[col])
                    print('boxcox trans')

                # yeojohn transform can be general for everything else
                else:
                    yeojohn_trans, lambda_ = yeojohnson(df[col])
                    print('yeojohnson trans')

                    
            trans_obj = [log_trans,boxcox_trans, yeojohn_trans]

            # obj is one of the trans_obj depend on condition to do the trans
            for i, obj in enumerate(trans_obj, start=1):
                if obj is not None: 
                    # store the trans col back to the original col position
                    df[col] = obj
                    statistic, p_value = stats.shapiro(obj)        
                    print("After transform for normality: statistic: ", statistic, "p_value: ", p_value)

                    plt.figure(figsize=(4, 2))  # Adjust the figsize as needed
                    stats.probplot(obj, dist="norm", plot=plt)
                    plt.title("Q-Q Plot for trans col: " + col)
                    plt.xlabel("Theoretical Quantiles")
                    plt.ylabel("Sample Quantiles")
                    plt.show()
                    # print(i, obj)
                    # plt.figure(figsize=(4,2))
                    # sns.histplot(obj, kde=True,  label='{i}: {obj}', stat='density')
                    # plt.title("Histogram of trans col: " + col)
                    # plt.xlabel("Values")
                    # plt.ylabel("Frequency")
                    # plt.show()
    return(df)
    # # for col in df
    
# df = handle_skewness(df)
```

# Normalization by L1 / L2
### scale features based on diff methematical norms (L1: obs/sum(abs(all features)) = 1, L2:obs/sqrt(sum(all features)

```python
# L1: Manhattan normalization, scales each feature by dividing each observation by the sum of the absolute values of all the features in the dataset
# It's particularly useful when you have sparse data or when you want to reduce the impact of outliers.
# L2: Euclidean normalization, scales each feature by dividing each observation by the square root of the sum of the squares of all the features in the dataset.
# It's useful when you want to minimize the impact of large values in your dataset.

 
from sklearn.preprocessing import Normalizer
import pandas as pd

# L1 Normalization
normalizer_l1 = Normalizer(norm='l1')
l1norm_array = normalizer_l1.fit_transform(df)
# Convert the normalized data back to a DataFrame
df_l1norm = pd.DataFrame(l1norm_array, columns=df.columns)

# L2 Normalization
normalizer_l2 = Normalizer(norm='l2')
l2norm_array = normalizer_l2.fit_transform(df)
# Convert the normalized data back to a DataFrame
df_l2norm = pd.DataFrame(l2norm_array, columns=df.columns)

# print("Original DataFrame:")
# print(df)
# print("\nL1 Normalized DataFrame:")
# print(normalized_df_l1)
# print("\nL2 Normalized DataFrame:")
# print(normalized_df_l2)
```

# Handle Multicollinearity
### becomes difficult to distinguish the individual effects of these correlated features on the target variable.

```python
# detect Multicollinearity
#############################################################################
# Compute correlation matrix: High correlations (close to 1 or -1) indicate potential multicollinearity.
# dataset where there are multiple features and a label y, using corrrelation matrix, we want to eliminate 
# some of the Xs' if X & X are highly correlated and they considered multicollinearity, but we also want to drop those for low correlation between 
# Xs' and y since low correlation of that means the Xs' don't contribution much predictor power to y or introduce noise 


corr_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
#############################################################################
# Calculate VIF
# VIF (Variance Inflation Factor): Calculate the VIF for each feature. VIF measures how much the variance of an estimated regression coefficient increases 
#if your predictors are correlated.
# Multicollinearity can inflate standard errors, leading to unstable and unreliable estimates of the regression coefficients.
# A VIF greater than 10 or 5 is often considered indicative of multicollinearity.
# A VIF greater than 10 or 5 is often considered indicative of multicollinearity.
X = np.array(df)
vif_data = pd.DataFrame()
vif_data["Feature"] = df.columns
vif_data["VIF"] = [variance_inflation_factor(X, i) for i in range(X.shape[1])]

print("Variance Inflation Factor (VIF):")
print(vif_data)
# # X_modified = modified_df[['independent_var1', 'independent_var2', ...]]
# vif_modified = pd.DataFrame()
# vif_modified["VIF Factor"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
# vif_modified["features"] = df.columns
# vif_modified

#############################################################################
# Eigenvalue of PCA: Large eigenvalues suggest multicollinearity.
cov_matrix = np.cov(df, rowvar=False)
eigenvalues, _ = eig(cov_matrix)
print('eigenvalues:' )
display(dict(zip(list(df.columns), eigenvalues)))
```

```python
###############################################################################################
# Recursive Feature Elimination (RFE):
# RFE works by recursively fitting the model and eliminating the least significant features based on their importance ranking.
# Initialize the model
# model = LinearRegression()
# model.fit(X, y)
# # Get the coefficient estimates
# coefficients = model.coef_.round(3)
# # assuming all features are used:
# # display(dict(zip(list(X.columns), coefficients)))
# coe_est = pd.DataFrame()
# coe_est["Feature"] = X.columns
# coe_est["coefficients"] = coefficients
# display(coe_est)
print("Recursive Feature Elimination (RFE)")
# Initialize RFE
model = LinearRegression()
rfe = RFE(model, n_features_to_select=3)
# Fit RFE
rfe.fit(X, y)

# rfe.support_: This attribute returns a boolean array indicating whether a feature is selected (True) or not (False). The selected features will have a value of True.s
# rfe.ranking_: This attribute returns the ranking of features. The feature with a ranking of 1 is considered the most important, followed by 2, and so on.
rfe_result = pd.DataFrame()
rfe_result["Feature"] = X.columns
rfe_result["support"] = rfe.support_
rfe_result["ranking"] = rfe.ranking_
display(rfe_result)
print("*"*70)

###############################################################################################
# Feature Importance from Trees:
# Tree-based models like Random Forest or Gradient Boosting can provide feature importance scores. 
#  scores represent the relative importance of each feature in predicting the target variable.
# Higher scores indicate that the feature is more important in making accurate predictions, while lower scores suggest less importance.
print("Feature Importance from Trees")
model = RandomForestRegressor()
# Fit the model
model.fit(X, y)
# Get feature importances
# Feature importance scores represent the relative importance of each feature in predicting the target variable. 
# Higher scores indicate that the feature is more important in making accurate predictions, while lower scores suggest less importance.
mdl_result = pd.DataFrame()
mdl_result["Feature"] = X.columns
mdl_result['feature_imp'] = model.feature_importances_
display(mdl_result)
print("*"*70)

###############################################################################################
# SelectKBest:
# This method selects the top k features based on statistical tests like ANOVA for classification and F-regression for regression.
# Initialize SelectKBest
print("SelectKBest:")
selector = SelectKBest(score_func=f_regression, k=3)
# Fit SelectKBest
X_new = selector.fit_transform(X, y)
selected_features_mask = selector.get_support()
selkbest_result = pd.DataFrame()
selkbest_result["Feature"] = X.columns
selkbest_result['feature_imp'] = selected_features_mask
display(selkbest_result)
print("*"*70)

###############################################################################################
# L1 Regularization (Lasso): (result still look very like the lm coefficient estimates
# Lasso regularization penalizes the absolute size of the coefficients, effectively setting some coefficients to zero and thus performing feature selection.
# This encourages the model to shrink less important feature coefficients to zero, effectively performing feature selection by excluding those features from the model.# The regularization strength (alpha) controls the degree of penalization applied to the coefficients. A higher alpha value leads to more coefficients being pushed towards zero, 
# It's important to evaluate the performance of the Lasso model, such as by examining metrics like mean squared error (MSE), R-squared, or cross-validated scores, to ensure that the selected features contribute to a predictive and generalizable model.
# large alpha not change result much: it suggests that the regularization effect of the Lasso is not strong enough to cause significant shrinkage or feature selection.
 # more effective when there are many irrelevant features or high multicollinearity among features. 
print("L1 Regularization (Lasso):") 
lasso = Lasso(alpha=0.5)
# Fit Lasso model
lasso.fit(X, y)
coefficients = lasso.coef_
lasso_result = pd.DataFrame()
lasso_result["Feature"] = X.columns
lasso_result['feature_imp'] = coefficients
display(lasso_result)
# X.columns[coefficients != 0]
print("*"*70)

###############################################################################################
# Recursive Feature Addition:Recursive Feature Elimination with Cross-Validation (RFECV) (not good for SMALL DF)
# This is similar to RFE but instead of eliminating features, it adds features one by one based on their importance.
print("Recursive Feature Addition:")
# Initialize the model
model = LinearRegression()
# Initialize RFECV
rfecv = RFECV(estimator=model, step=1, scoring='neg_mean_squared_error')s
# Fit RFECV
rfecv.fit(X, y)
feature_ranking = rfecv.ranking_
selected_features_mask = rfecv.support_
cross_val_scores = rfecv.cv_results_['mean_test_score']
optimal_num_features = rfecv.n_features_
rfecv_result = pd.DataFrame()
rfecv_result["Feature"] = X.columns
rfecv_result["support"] = selected_features_mask
rfecv_result["ranking"] = feature_ranking
rfecv_result["cv_score"] = cross_val_scores
rfecv_result["n_features"] = optimal_num_features
display(rfecv_result)

print("*"*70)


```

## Factor Analysis of Mixed Data (FAMD)
### unsupervised for both numerical & categoricals

```python
# FAMD is a technique similar to PCA but specifically designed for datasets with both numerical and categorical variables.
# It can be used for feature selection in unsupervised learning scenarios.
# row contribution: Higher contributions indicate that a particular observation has a greater influence on defining the factors.
# column conntribution: Higher contributions indicate that a particular feature has a greater influence on defining the factors.
# Higher eigenvalues indicate that the corresponding factor captures more variance in the data and is more important in defining the factor space.
# Initialize FAMD
famd = FAMD(n_components=5)
# Fit FAMD
famd.fit(X)
print("Eigenvalues:")
display(famd.eigenvalues_summary)
print("Row coordinates:")
display(famd.row_coordinates(X).head())
print("Column coordinates:")
display(famd.column_coordinates_)
print("Row Contributions:")
display((
    famd.row_contributions_
    .sort_values(0, ascending=False)
    .head(5)
    .style.format('{:.3%}')))
print("Column Contributions;")
display(famd.column_contributions_.style.format('{:.0%}'))
# top components take up most of the variances of the data
famd.plot(
    X,
    x_component=0,
    y_component=1
)
```

```python
from sklearn.preprocessing import LabelEncoder

# Assuming X contains your feature set with categorical variables
X_encoded = X.copy()  # Make a copy of X to avoid modifying original data
label_encoders = {}

for column in X_encoded.columns:
    if X_encoded[column].dtype == 'object':
        label_encoders[column] = LabelEncoder()
        X_encoded[column] = label_encoders[column].fit_transform(X_encoded[column])

# Now X_encoded contains numerical representations of categorical variables
# Proceed with applying SMOTE

```

```python
from imblearn.under_sampling import RandomUnderSampler

# Undersampling: Randomly remove samples from the majority class to balance the dataset.
# remove till each class level obs reach to the min of original data class level
# Undersample the majority class
# resample not required Xs to be numerical (apply for any X data)
undersampler = RandomUnderSampler()
X_resampled, y_resampled = undersampler.fit_resample(X, y)

print('total sample size after Undersample:', len(y_resampled))
display(y_resampled.value_counts())
display(y_resampled.value_counts(normalize = True))
```

```python
# Oversampling: Randomly duplicate samples from the minority class to balance the dataset.
# oversample till each class level obs reach to the max of original data class level
# resample not required Xs to be numerical (apply for any X data)
from imblearn.over_sampling import RandomOverSampler
# Oversample the minority class
oversampler = RandomOverSampler()
X_resampled, y_resampled = oversampler.fit_resample(X, y)

print('total sample size after Oversampling:', len(y_resampled))
display(y_resampled.value_counts())
display(y_resampled.value_counts(normalize = True))
```

```python
# SMOTE (Synthetic Minority Over-sampling Technique): Generate synthetic samples for the minority class.
# resample required Xs to be numerical (apply for numerical X data)

from imblearn.over_sampling import SMOTE
# Apply SMOTE
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X_encoded, y)

print('total sample size after SMOTE:', len(y_resampled))
display(y_resampled.value_counts())
display(y_resampled.value_counts(normalize = True))
```
